Debes realizar un an√°lisis que responda a las siguientes preguntas, usando f√≥rmulas geom√©tricas para el c√°lculo de √°reas:

- ¬øCu√°l es el √°rea total del terreno disponible?

- ¬øQu√© pasa si se aproxima el n√∫mero de m√≥dulos por exceso o por defecto?
- ¬øCu√°l ser√≠a la mejor decisi√≥n para aprovechar al m√°ximo el espacio sin comprometer el despliegue?

üìí*Presenta un informe de salida donde se explique por qu√© se eligi√≥ esa opci√≥n como la m√°s adecuada.*

````python
import math

# Captura de datos
# Datos M√≥dulo Solar - Forma Trapezoidal
mod_solar_b_mayor = 10
mod_solar_b_menor = 4
mod_solar_altura = 4
# Datos Terreno Disponible - Forma Rombo
terreno_diag_mayor = 30
terreno_diag_menor = 16

### C√°lculos
# √Åreas
area_modulo = ((mod_solar_b_mayor + mod_solar_b_menor) * mod_solar_altura)/2
area_lote = (terreno_diag_mayor * terreno_diag_menor)/2
# Cantidades
cant_modulos = area_lote / area_modulo
cant_mod_completos = math.floor(cant_modulos) # M√≥dulos que caben completos
cant_mod_excesos = math.ceil(cant_modulos)    # M√≥dulos incluso que se salen
# √Årea de Exceso
area_modulos_exceso = cant_mod_excesos * area_modulo # Los m√≥dulos por el √°rea de cada uno
area_completa_exceso = area_modulos_exceso - area_lote #  √Årea que se excede menos el lote
# Porcentaje de Exceso
porc_exceso = ( area_completa_exceso / area_lote) * 100

# Visualizaci√≥n
print(f'''
En un lote de Dimensiones:       {terreno_diag_mayor} m¬≤ x {terreno_diag_menor} m¬≤
√Årea de cada M√≥dulo:             {area_modulo} m¬≤
Cantidad de M√≥dulos con Decimal: {cant_modulos}
Cantidad M√≥dulos Completos:      {cant_mod_completos}
Cantidad M√≥dulos con Exceso:     {cant_mod_excesos}
_______________________________________________
√Årea Exceso:                     {area_completa_exceso} m¬≤
Porcentaje de Exceso:            {porc_exceso} %

''')
````


<font size=7 color=  F26DCD>üéÄ **Mini-reto 4** | Trazando Caminos Sostenibles en el Territorio

<font size=5 color= F1CFFF>Contexto del Problema  

En el dise√±o de rutas para la conectividad entre comunidades rurales, es fundamental conocer la **distancia real entre dos puntos** del territorio. Ya sea para proyectar senderos peatonales, l√≠neas de distribuci√≥n de energ√≠a o recorridos escolares, calcular la distancia en un plano cartesiano es el primer paso para una planificaci√≥n territorial eficiente.

Los datos de entrada son coordenadas cartesianas de dos puntos:  
- Punto de origen: *(x‚ÇÅ, y‚ÇÅ)*  
- Punto de destino: *(x‚ÇÇ, y‚ÇÇ)*  

Se requiere calcular la distancia entre ambos puntos usando la f√≥rmula de la **distancia euclidiana**.

<font size=5 color= F1CFFF>Objetivos  
- Comprender y aplicar la f√≥rmula de la distancia entre dos puntos.  
- Desarrollar un programa en Python que:  
  - Solicite al usuario las coordenadas de dos puntos.  
  - Calcule la distancia entre ellos.  
  - Muestre el resultado redondeado a dos decimales.

<font size=5 color= F1CFFF>Criterios de Calidad  
- ‚úÖ El c√°lculo debe ser matem√°ticamente correcto.  
- ‚úÖ Las entradas del usuario deben validarse como num√©ricas.

<font size=5 color= F1CFFF> Pistas para los Campistas  
- La f√≥rmula de la distancia entre dos puntos es:

  $$
  d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}
  $$

- Importa la librer√≠a `math` al inicio con `import math`.  
- Usa `input()` para capturar datos del usuario y `float()` para convertirlos.  
- Usa `math.sqrt()` para la ra√≠z cuadrada y `math.pow()` para las potencias.  
- Puedes mostrar el resultado as√≠:  
  ```python
  print(f"La distancia es: {distancia:.2f} unidades")
  ```

> üåç *Este reto nos prepara para manejar coordenadas espaciales, un elemento clave en sistemas de informaci√≥n geogr√°fica y aplicaciones de planeaci√≥n territorial.*


````python
import math
# Resuelve el problema en esta celda

### Captura de Datos
x_1 = float(input('Ingrese el valor de x‚ÇÅ ‚Üí'))
y_1 = float(input('Ingrese el valor de y‚ÇÅ ‚Üí'))
x_2 = float(input('Ingrese el valor de x‚ÇÇ ‚Üí'))
y_2 = float(input('Ingrese el valor de y‚ÇÇ ‚Üí'))

### C√°lculos
desp_hor = x_2 - x_1
desp_ver = y_2 - y_1

discrim = math.pow(desp_hor, 2) + math.pow(desp_ver, 2)
distancia = math.sqrt(discrim)

### Visualizaci√≥n
print(f'''
La distancia entre ({x_1}, {y_1}) y ({x_2}, {y_2}) ser√≠a {distancia:.2f} metros
''')
````

En el dise√±o de sistemas sostenibles para la **gesti√≥n de aguas lluvias**, una comunidad ha optado por construir canales con una **pendiente parab√≥lica**, ya que esta forma mejora la eficiencia del flujo sin causar erosi√≥n.

La trayectoria del flujo del agua se modela con una funci√≥n cuadr√°tica del tipo:

$$
y = ax^2 + bx + c
$$

Para conocer los puntos exactos donde el canal toca el terreno (es decir, cruza el eje horizontal), se requiere **resolver las ra√≠ces de la ecuaci√≥n cuadr√°tica**.  
Este c√°lculo permitir√° determinar con precisi√≥n el **inicio y el final del canal**, informaci√≥n clave para proyectar la obra civil y calcular su longitud efectiva.

<font size=5 color= 657BC7>Objetivos  
- Implementar en Python una soluci√≥n que:
  - Solicite al usuario los coeficientes *a*, *b* y *c*.
  - Calcule las soluciones reales (si existen) usando la f√≥rmula general:
  
    $$
    x = \frac{- b \pm \sqrt{b^2 - 4ac}}{2a}
    $$

  - Analice el discriminante para indicar si hay dos soluciones, una √∫nica soluci√≥n o ninguna soluci√≥n real.
  - Presente los resultados de forma clara.

<font size=5 color= 657BC7>Criterios de Calidad  
- ‚úÖ El c√≥digo debe manejar adecuadamente los diferentes casos seg√∫n el valor del discriminante (*positivo, cero o negativo*).  
- ‚úÖ El resultado debe mostrarse con precisi√≥n y lenguaje interpretativo.  
- ‚úÖ Validaci√≥n b√°sica para evitar divisi√≥n por cero cuando *a = 0*.  
- ‚úÖ El c√≥digo debe estar correctamente comentado y estructurado.  
- ‚úÖ Se debe redondear el resultado a dos decimales para mejor legibilidad.

<font size=5 color= 657BC7>Pistas para los Campistas  
- El **discriminante** es la parte dentro de la ra√≠z cuadrada:  
  $$
  D = b^2 - 4ac
  $$

  - Si *D > 0*: hay **dos soluciones reales distintas**.  
  - Si *D = 0*: hay **una soluci√≥n real doble**.  
  - Si *D < 0*: no hay soluci√≥n real (ra√≠ces imaginarias).

- Usa `math.sqrt()` para calcular la ra√≠z cuadrada del discriminante.  
- Controla el caso en que `a == 0`, ya que la f√≥rmula no ser√≠a v√°lida (no es cuadr√°tica).  
- Utiliza `f-strings` para mostrar los resultados con dos decimales.

> üßÆ *Este reto fortalece tus habilidades para resolver ecuaciones con significado en proyectos reales, combinando matem√°tica con sostenibilidad y planificaci√≥n territorial.*

````python
import math

### Captura de Datos
a = float(input('Ingrese el coeficiente de x¬≤ '))
b = float(input('Ingrese el coeficiente de x '))
c = float(input('Ingrese el t√©rmino independiente '))

discriminante = math.pow(b, 2) - 4*a*c
if a == 0:
    mensaje = 'Error de Divisi√≥n por Cero'
elif discriminante <0 :
    mensaje = 'No tiene Soluci√≥n en los Reales'
else:
    raiz = math.sqrt(discriminante)
    x_1 = (-b + raiz)/(2*a)
    x_2 = (-b - raiz)/(2*a)
    mensaje = 'Ok'

if mensaje == 'Ok':
    print(f'''
    El valor de x‚ÇÅ es: {x_1:.2f}
    El valor de x‚ÇÇ es: {x_2:.2f}
    ''')
else:
    print(mensaje)
````

## Merge - Fusi√≥n

```python
merge
```

Es parecido a unir tablas en **SQL**. Con la Fusi√≥n podemos **combinar filas de <u>dos o mas Dataframes</u>** en funci√≥n de una clave/claves(s) com√∫n(es). Por ejemplo, si tenemos un dataframe de productos y otro Dataframe de proveedores, podemos fuionarlos en un nuevo Dataframe que nos diga que proveedor suministra un producto.

````python
df_detalle = df_material_proyecto.merge(df_proyecto, on = 'IdProyecto').merge(df_material, on = 'IdMaterial')
display(df_detalle)
````

````python
### Filtrar solo las columnas de
df_detalle_proy_mat = df_detalle[['Nombre Proyecto', 'Nombre Material', 'Cantidad']]
display(df_detalle_proy_mat)
````

````python
### Calcular reporte de materiales (COLUMNS) por Nombre de Proyecto (INDEX)
df_pivote_proyectos = df_detalle.pivot_table(
    index='Nombre Proyecto',
    columns='Nombre Material',
    values='Cantidad',
    aggfunc='sum'
)
print('\n','_'*60)
df_pivote_proyectos = df_pivote_proyectos.fillna(0)
display(df_pivote_proyectos)
````
